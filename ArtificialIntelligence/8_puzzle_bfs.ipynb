```python
#8 puzzle with BFS
import copy

initial_state=[[1,2,3],[8,' ',4],[7,6,5]]
goal_state=[[2,8,1],[' ',4,3],[7,6,5]]
open=[[initial_state,None]]
closed=[]
visited=[]

def bfs(open,closed):
    while open:
        temp=open.pop(0)
        closed.append(temp)
        if temp[0] not in visited:
            visited.append(temp[0])
        if temp[0]==goal_state:
            pathgenerator(closed,temp[1],temp[0])
            print('success')
            # print(closed)
            return
        else:
            next=movegen(temp[0])
            for i in next:
                if i not in visited:
                    open.append([i,temp[0]])
                    visited.append(i)
def pathgenerator(closed,parent,goal):
    if parent==None:
        print('->',goal)
        return
    for i in closed:
        if i[0]==goal:
            for j in closed:
                if j[0]==i[1]:
                    pathgenerator(closed,j[1],j[0])
                    print('->',i[0])
                    return    

def movegen(curr):
    c=copy.deepcopy(curr)
    next=[]
    bi=blankindex(curr)
    if(bi[0]!=0):
        next.append(moveup(c,bi))
        c=copy.deepcopy(curr)
    if(bi[0]!=2):
        next.append(movedown(c,bi))
        c=copy.deepcopy(curr)
    if(bi[1]!=0):
        next.append(moveleft(c,bi))
        c=copy.deepcopy(curr)
    if(bi[1]!=2):
        next.append(moveright(c,bi))
    return(next)

def blankindex(curr):
    count=0
    for i in curr:
       if ' ' in i:
           return(count,i.index(' '))
       count=count+1
        
def moveup(curr,bi):
    temp1=curr.copy()
    tempele1=temp1[bi[0]-1][bi[1]]
    temp1[bi[0]-1][bi[1]]=' '
    temp1[bi[0]][bi[1]]=tempele1
    return temp1
def movedown(curr,bi):
    temp2=curr.copy()
    tempele2=temp2[bi[0]+1][bi[1]]
    temp2[bi[0]+1][bi[1]]=' '
    temp2[bi[0]][bi[1]]=tempele2
    return temp2
def moveleft(curr,bi):
    temp3=curr.copy()
    tempele3=temp3[bi[0]][bi[1]-1]
    temp3[bi[0]][bi[1]-1]=' '
    temp3[bi[0]][bi[1]]=tempele3
    return temp3
def moveright(curr,bi):
    temp4=curr.copy()
    tempele4=temp4[bi[0]][bi[1]+1]
    temp4[bi[0]][bi[1]+1]=' '
    temp4[bi[0]][bi[1]]=tempele4
    return temp4

bfs(open,closed)
```
