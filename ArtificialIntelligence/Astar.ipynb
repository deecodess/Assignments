```python
#A*
import heapq
import math

class Graph:
    def __init__(self, edges):
        self.edges = edges
        self.nodes = set(sum(([edge[0], edge[1]] for edge in self.edges), []))

    def neighbors(self, node):
        return set([edge[1] for edge in self.edges if edge[0] == node])

    def cost(self, from_node, to_node):
        for edge in self.edges:
            if from_node == edge[0] and to_node == edge[1]:
                return edge[2]
        return math.inf

def astar(graph, start, goal):
    pq = [(0, start)]
    visited = set()
    came_from = {}
    cost_so_far = {}
    came_from[start] = None
    cost_so_far[start] = 0

    while pq:
        _, current = heapq.heappop(pq)
        if current == goal:
            break
        visited.add(current)

        for next in graph.neighbors(current):
            new_cost = cost_so_far[current] + graph.cost(current, next)
            if next not in cost_so_far or new_cost < cost_so_far[next]:
                cost_so_far[next] = new_cost
                priority = new_cost + heuristic(next, goal)
                heapq.heappush(pq, (priority, next))
                came_from[next] = current

    path = []
    current = goal
    while current != start:
        path.append(current)
        current = came_from[current]
    path.append(start)
    path.reverse()

    return path, cost_so_far[goal]

def heuristic(node, goal):
    x1 = ord(node) - 65
    y1 = 0
    x2 = ord(goal) - 65
    y2 = 0
    return math.sqrt((x1 - x2)**2 + (y1 - y2)**2)

# Define the graph
graph = Graph([
    ('A', 'B', 2),
    ('A', 'C', 4),
    ('B', 'D', 5),
    ('B', 'E', 12),
    ('C', 'F', 3),
    ('D', 'G', 6),
    ('E', 'F', 7),
    ('E', 'H', 10),
    ('F', 'H', 2),
    ('G', 'H', 3)
])

# Define the start and goal nodes
start = 'A'
goal = 'H'

# Run the A* algorithm on the graph
path,cost = astar(graph, start, goal)

# Print the results
print('Path:', path)
print('Cost:', cost)
```
