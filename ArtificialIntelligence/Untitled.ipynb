```python
import heapq, copy

def up(initial_state):
  temp = copy.deepcopy(initial_state)
  i0 = -1
  j0 = -1
  for i in range(3):
    for j in range(3):
      if temp[i][j]==0:
        i0 = i
        j0 = j 
        break
  if i0>0:
    temp[i0][j0] = temp[i0-1][j0]
    temp[i0-1][j0] = 0
    return temp
  return []

def down(initial_state):
  temp = copy.deepcopy(initial_state)
  i0 = -1
  j0 = -1
  for i in range(3):
    for j in range(3):
      if temp[i][j]==0:
        i0 = i
        j0 = j 
        break
  if i0<2:
    temp[i0][j0] = temp[i0+1][j0]
    temp[i0+1][j0] = 0
    return temp
  return []

def left(initial_state):
  temp = copy.deepcopy(initial_state)
  i0 = -1
  j0 = -1
  for i in range(3):
    for j in range(3):
      if temp[i][j]==0:
        i0 = i
        j0 = j 
        break
  if j0>0:
    temp[i0][j0] = temp[i0][j0-1]
    temp[i0][j0-1] = 0
    return temp
  return []

def right(initial_state):
  temp = copy.deepcopy(initial_state)
  i0 = -1
  j0 = -1
  for i in range(3):
    for j in range(3):
      if temp[i][j]==0:
        i0 = i
        j0 = j 
        break
  if j0<2:
    temp[i0][j0] = temp[i0][j0+1]
    temp[i0][j0+1] = 0
    return temp
  return []



close = []

adjl = {
    'A': [['B'], ['C','D']],
    'B': [['E'], ['F']],
    'C': [],
    'D': [['G', 'H']],
    'E': [],
    'F': []
}

h = {
     'A': -1,
     'B': 6,
     'C': 12,
     'D': 10,
     'E': 5,
     'F': 7,
     'G': 4,
     'H': 4
}

explored = {
    'A': False,
    'B': False,
    'C': False,
    'D': False,
    'E': False,
    'F': False,
    'G': False,
    'H': False
}

min_next = {
    'A': [],
    'B': [],
    'C': [],
    'D': [],
    'E': [],
    'F': [],
    'G': [],
    'H': []
}

def reset_explored():
  for i in ['A', 'B', 'C', 'D', 'E', 'F', 'G', 'H']:
    explored[i] = False

def solve(curr_state):
  open = []
  for i in adjl[curr_state]:
    cost = 0
    for j in i:
      cost += h[j] + 1
    heapq.heappush(open, [cost, i])
  if len(open)==0:
    return
  prev_cost = -1
  while True:
    next_cost, next_states = heapq.heappop(open)
    if prev_cost==next_cost:
      heapq.heappush(open, [next_cost, next_states])
      break
    temp_cost = 0
    for state in next_states:
      if not explored[state]:
        solve(state)
      temp_cost += h[state] + 1
    heapq.heappush(open, [temp_cost, next_states])
    prev_cost = next_cost
  min_cost, min_child = heapq.heappop(open)
  h[curr_state] = min_cost
  min_next[curr_state] = min_child
  reset_explored()

solve('A')
print(h)
print(min_next)
```
